```python
import uiat # it will takes some time.
```

### parameter setting (default)

```python
n = 1000 #changeable (sample size)
c1 = 1
c2 = 1 
c3 = 1 #changeable (the strength of confounding)
c4 = 1
c5 = 1
r1 = r2 = 1
r3 = r4 = 1 #changeable (the strength of correlation in setting 5 and 6)

import scipy.stats as stats 
import numpy as np
import pandas as pd
```

### Simulation

```python
#Setting 1
np.random.seed(1)

COUNT1=0 # number of rejection
for i in range(1000):

    U = np.random.normal(0,1,n) #common cause of X1 and X2
    e1 = np.random.normal(0,1,n)
    e2 = np.random.normal(0,1,n)
    ey = np.random.normal(0,1,n)
    
    X1 = r1*U+e1
    X2 = r2*U+e2
    Y = c1*X1+ey
    
    if i==1:
        f1_cor12=stats.pearsonr(X1, X2)[0]
    
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,dtype="continuous",verbose=False)
    pvalue = UIAT.test()
    if pvalue<=0.05:
        COUNT1 = COUNT1+1 
```

```python
#Setting 2
np.random.seed(1)

COUNT21=0 # number of rejection when confounder is considered.
COUNT22=0 # number of rejection when confounder is not considered.
for i in range(1000):
    U = np.random.normal(0,1,n) #common cause of X1 and X2
    C = np.random.normal(0,1,n) #Confounder
    e1 = np.random.normal(0,1,n)
    e2 = np.random.normal(0,1,n)
    ey = np.random.normal(0,1,n)
    
    X1 = r1*U+c2*C+e1
    X2 = r2*U+e2
    Y = c1*X1+c3*C+ey
    
    if i==1:
        f2_cor12=stats.pearsonr(X1, X2)[0]
   
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,covariates=C,dtype="continuous",verbose=False)
    pvalue1 = UIAT.test()
    if pvalue1<=0.05:
        COUNT21 = COUNT21+1
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,dtype="continuous",verbose=False)
    pvalue2 = UIAT.test()
    if pvalue2<=0.05:
        COUNT22 = COUNT22+1 
```

```python
#Setting 3
np.random.seed(1)

COUNT31=0 # number of rejection when confounder is considered.
COUNT32=0 # number of rejection when confounder is not considered.
for i in range(1000):
    U = np.random.normal(0,1,n) #common cause of X1 and X2
    C = np.random.normal(0,1,n) #Confounder
    e1 = np.random.normal(0,1,n)
    e2 = np.random.normal(0,1,n)
    ey = np.random.normal(0,1,n)
    
    X1 = r1*U+c2*C+e1
    X2 = r2*U+e2
    Y = c3*C+ey
    
    if i==1:
        f3_cor12=stats.pearsonr(X1, X2)[0]
        
        
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,covariates=C,dtype="continuous",verbose=False)
    pvalue1 = UIAT.test()
    if pvalue1<=0.05:
        COUNT31 = COUNT31+1 
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,dtype="continuous",verbose=False)
    pvalue2 = UIAT.test()
    if pvalue2<=0.05:
        COUNT32 = COUNT32+1 
```

```python
#Setting 4 
np.random.seed(1)

COUNT41=0 # number of rejection when confounder is considered.
COUNT42=0 # number of rejection when confounder is not considered.
for i in range(1000):
    U = np.random.normal(0,1,n) #common cause of X1 and X2 and X3
    C = np.random.normal(0,1,n) #Confounder
    e1 = np.random.normal(0,1,n)
    e2 = np.random.normal(0,1,n)
    e3 = np.random.normal(0,1,n)
    ey = np.random.normal(0,1,n)
    
    X1 = r1*U+c2*C+e1
    X2 = r2*U+e2
    X3 = r3*U+e3
    Y = c1*X1+c4*X2+c3*C+ey
    
    if i==1:
        f4_cor13=stats.pearsonr(X1, X3)[0]
        
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X3,covariates=np.column_stack((X2,C)),dtype="continuous",verbose=False)
    pvalue1 = UIAT.test()
    if pvalue1<=0.05:
        COUNT41 = COUNT41+1 
        
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X3,covariates=X2,dtype="continuous",verbose=False)
    pvalue2 = UIAT.test()
    if pvalue2<=0.05:
        COUNT42 = COUNT42+1 
```

```python
#Setting 5
np.random.seed(1)

COUNT51=0 # number of rejection when confounder is considered. (for EV X2)
COUNT52=0 # number of rejection when confounder is not considered. (for EV X2)
COUNT53=0 # number of rejection when confounder is considered. (for EV X3)
COUNT54=0 # number of rejection when confounder is not considered. (for EV X3)
for i in range(1000):
    U = np.random.normal(0,1,n) #common cause of X1 and X2 and X3
    C = np.random.normal(0,1,n) #Confounder
    e1 = np.random.normal(0,1,n)
    e2 = np.random.normal(0,1,n)
    e3 = np.random.normal(0,1,n)
    ey = np.random.normal(0,1,n)
    
    X1 = r1*U+c2*C+e1
    X2 = r2*U+e2
    X3 = r3*U+e3
    Y = c1*X1+c3*C+ey
    
    if i==1:
        f5_cor12=r1*r2/np.sqrt(r1**2+c2**2+1)/np.sqrt(r2**2+1)
        f5_cor13=r1*r3/np.sqrt(r1**2+c2**2+1)/np.sqrt(r3**2+1)

        
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,covariates=C,dtype="continuous",verbose=False)
    pvalue1 = UIAT.test()
    if pvalue1<=0.05:
        COUNT51 = COUNT51+1 
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X2,dtype="continuous",verbose=False)
    pvalue2 = UIAT.test()
    if pvalue2<=0.05:
        COUNT52 = COUNT52+1 
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X3,covariates=C,dtype="continuous",verbose=False)
    pvalue3 = UIAT.test()
    if pvalue3<=0.05:
        COUNT53 = COUNT53+1 
    UIAT = uiat.UniversalIgnorabilityAssumptionTest(cause=X1,effect=Y,explorer=X3,dtype="continuous",verbose=False)
    pvalue4 = UIAT.test()
    if pvalue4<=0.05:
        COUNT54 = COUNT54+1 
```
